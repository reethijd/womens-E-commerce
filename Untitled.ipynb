```python
import pickle

class TextPreprocessor:
    def preprocess_text(self, text):
        text = text.lower()
        text = re.sub(r'[^a-zA-Z\s]', '', text)
        tokens = text.split()
        tokens = [word for word in tokens if word not in stopwords.words('english')]
        stemmer = PorterStemmer()
        tokens = [stemmer.stem(word) for word in tokens]
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(word) for word in tokens]
        return ' '.join(tokens)

# Save the class instance
preprocessor = TextPreprocessor()
with open('text_preprocessor.pkl', 'wb') as f:
    pickle.dump(preprocessor, f)
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from gensim import corpora
from gensim.models import LdaModel

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Load DataFrame
df = pd.read_csv("E-commerce")
text_data = df['Review Text'].apply(lambda x: x.lower().split())

# Preprocess
def preprocess(text):
    tokens = [word for word in text if word.isalpha() and word not in stop_words]
    return ' '.join(tokens)

text_data = text_data.apply(preprocess)

# Vectorization
vectorizer = CountVectorizer()
dtm = vectorizer.fit_transform(text_data)

# Dictionary and Corpus
dictionary = corpora.Dictionary([text.split() for text in text_data])
corpus = [dictionary.doc2bow(text.split()) for text in text_data]

# LDA Model
lda_model = LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15)

# Print Topics
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}: {topic}")

# Assign topics
topic_assignments = lda_model.get_document_topics(corpus)
df['topic'] = [max(doc, key=lambda x: x[1])[0] for doc in topic_assignments]
```

```python
# -------------------- Bagging with LSTM --------------------
def create_lstm_model():
    model = Sequential()
    model.add(Embedding(input_dim=max_words, output_dim=128, input_length=max_len))
    model.add(LSTM(units=128, return_sequences=False))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Step 1: Create an empty list to store predictions from each LSTM model
all_predictions = []

# Step 2: Create multiple LSTM models
n_estimators = 5  # Number of LSTM models to create
for _ in range(n_estimators):
    lstm_model = create_lstm_model()
    lstm_model.fit(X_train_pad, y_train, epochs=5, batch_size=64, verbose=0)
    
    # Step 3: Make predictions on the test set
    y_pred_prob = lstm_model.predict(X_test_pad)
    y_pred = (y_pred_prob >= 0.5).astype(int)
    all_predictions.append(y_pred)

# Step 4: Average the predictions from all models
y_pred_ensemble = np.mean(all_predictions, axis=0).round().astype(int)

# Step 5: Evaluate the ensemble predictions
accuracy_lstm = accuracy_score(y_test, y_pred_ensemble)
print("LSTM Bagging Accuracy:", accuracy_lstm)
print("LSTM Bagging Classification Report:")
print(classification_report(y_test, y_pred_ensemble))

# -------------------- Summary of Results --------------------
print("\nSummary of Bagging Results:")
print(f"Logistic Regression Bagging Accuracy: {accuracy_lr:.2f}")
print(f"SVM Bagging Accuracy: {accuracy_svm:.2f}")
print(f"LSTM Bagging Accuracy: {accuracy_lstm:.2f}")
```

```python
from flask import Flask, request, jsonify
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle
import traceback  # Import traceback for detailed error messages

# Load your saved LSTM model
model = tf.keras.models.load_model('lstm_model.h5')
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Load your tokenizer
with open('text_preprocessor.pkl', 'rb') as handle:
    tokenizer = pickle.load(handle)

# Set max_len
max_len = 100

# Initialize the Flask app
app = Flask(__name__)

# Define the API route for prediction
@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Get the input data from the request
        data = request.get_json(force=True)

        # Extract the 'text' from the input JSON
        input_text = data['text']
        
        # Check if input_text is empty
        if not input_text:
            return jsonify({'error': 'Input text cannot be empty.'}), 400

        # Preprocess the text (tokenize and pad)
        input_seq = tokenizer.texts_to_sequences([input_text])
        input_pad = pad_sequences(input_seq, maxlen=max_len)
        
        # Make the prediction
        pred_prob = model.predict(input_pad)[0][0]
        prediction = 1 if pred_prob >= 0.5 else 0
        
        # Send the response
        response = {
            'prediction': int(prediction),
            'probability': float(pred_prob)
        }
        return jsonify(response)
    
    except Exception as e:
        # Return detailed error information
        return jsonify({'error': str(e), 'traceback': traceback.format_exc()}), 500

# Run the Flask app
if __name__ == '__main__':
    app.run(debug=True)




```
